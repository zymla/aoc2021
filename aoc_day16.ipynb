```python
def parse():
    bit_count=0
    v=hex(int(''.join([input.pop(0) for i in range(3)]), 2))[2:]
    t=hex(int(''.join([input.pop(0) for i in range(3)]), 2))[2:]
    bit_count+=6
    print('version', v, 'type', t)
    if t=='4': # Literal value
        literal_list=[]
        while len(input):
            last_group=input.pop(0)=='0'
            bit_count+=1
            literal_list.extend([input.pop(0) for i in range(4)])
            bit_count+=4
            if last_group:
                break
        literal_bin=''.join(literal_list)
        literal_dec=int(literal_bin, 2)
        print('LITERAL', literal_bin, literal_dec, sep='\t')
        return (bit_count, [v], literal_dec)
    else: # Operator
        length_type=input.pop(0)
        bit_count+=1
        if length_type=='0':
            subpackets_bit_length_bin=''.join([input.pop(0) for i in range(15)])
            bit_count+=15
            subpackets_bit_length=int(subpackets_bit_length_bin, 2)
            print('SUBPACKETS', 'bit length', len(subpackets_bit_length_bin), subpackets_bit_length_bin, subpackets_bit_length, sep='\t')
            subpackets_total_bit_count=0
            vl=[]
            while subpackets_total_bit_count < subpackets_bit_length:
                sub_packetbit_count, v_subpacket, return_value = parse()
                vl.extend(v_subpacket)
                subpackets_total_bit_count+=sub_packetbit_count
            return (subpackets_total_bit_count+bit_count, [v]+vl, '')
        else:
            subpackets_nb_bin=''.join([input.pop(0) for i in range(11)])
            bit_count+=11
            subpackets_nb=int(subpackets_nb_bin, 2)
            print('SUBPACKETS', 'nb', len(subpackets_nb_bin), subpackets_nb_bin, subpackets_nb, sep='\t')
            subpackets_total_bit_count=0
            vl=[]
            for _ in range(subpackets_nb):
                sub_packetbit_count, v_subpacket, return_value = parse()
                vl.extend(v_subpacket)
                subpackets_total_bit_count+=sub_packetbit_count
            return (subpackets_total_bit_count+bit_count, [v]+vl, '')

```
